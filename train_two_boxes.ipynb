```python
import torch
from gym.wrappers.time_limit import TimeLimit
from boxforage.environments import BoxForagingEnvironment

env = BoxForagingEnvironment(boxes_spec={'num_boxes': 2, 'p_appear': 0.2, 'p_vanish': 0.1, 'p_true': 0.6, 'p_false': 0.4})
env = TimeLimit(env, max_episode_steps=40)

from stable_baselines3 import PPO

algo = PPO('MlpPolicy', env, n_steps=32, batch_size=16, device='cpu', verbose=1)
for c_idx in range(32):
    algo.learn(total_timesteps=320, log_interval=5, reset_num_timesteps=False)
    save_pth = 'store/p2164/ckpt_{:02d}.pt'.format(c_idx)
    torch.save(algo.policy.state_dict(), save_pth)
```
